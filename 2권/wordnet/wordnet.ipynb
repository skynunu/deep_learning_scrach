```python
import nltk
from nltk.corpus import wordnet
nltk.download('wordnet')

print("car 단어의 여러가지 뜻", )
print(wordnet.synsets('car')," \n")

print("car의 첫번째 의미의 정의")
car_n1 = wordnet.synset('car.n.01')
print(car_n1.definition()," \n")

print("car와 동의어들")
print(car_n1.lemma_names()," \n")

print("car와 다른 단어의 의미적 상하 관계")
car_n1.hypernym_paths()[0]
```

```python
#wordnet을 사용한 의미 유사도
car = wordnet.synset('car.n.01')
novel = wordnet.synset('novel.n.01')
dog = wordnet.synset('dog.n.01')
motorcycle = wordnet.synset('motorcycle.n.01')

print("car와 novel의 유사도:",car.path_similarity(novel))
print("car와 dog의 유사도:",car.path_similarity(dog))
print("car와 motorcycle의 유사도:",car.path_similarity(motorcycle))
```
